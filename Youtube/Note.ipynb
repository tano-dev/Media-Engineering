# Youtube Algorithm




### 1. The Core Philosophy: "The Matchmaker"

Forget the idea that the algorithm judges quality. It predicts **satisfaction**.

* **The Goal:** It wants to predict, with high certainty, that if it shows Video X to User Y, User Y will watch it, enjoy it, and *stay on YouTube*.
* **The Mechanism:**
* **The Hook (CTR):** The algorithm tests your video with a small "seed audience." If they click, it passes the first test.
* **The Hold (Retention):** If they click but leave in 30 seconds, the algorithm kills the video's reach. It assumes the title was "clickbait."
* **The Satisfaction (Session Time):** This is the most powerful metric in 2026. If a viewer watches your video and then *continues watching other videos* (even from other creators), your video gets credit for starting a "session."



### 2. Strategic Advantage: How to Win

You can manipulate these metrics by how you structure your content.

* **The "Session Time" Multiplier:**
* *Tactic:* Never end a video with "Thanks for watching, bye." Instead, say, "If you liked this, you'll want to see how we applied this logic to [Next Topic] in this video," and point to an end-screen element. This creates a "binge loop."


* **The "Search to Browse" Pipeline:**
* *Tactic:* New channels often grow through Search (solving a specific problem). Once you hook them with a solution, you convert them to "Browse" viewers (fans who watch whatever you post) by having a strong personality or unique editing style.



### 3. Tailored Content Niches for You

Based on your skillset, here are three high-potential paths that align with what the algorithm currently favors.

#### **Path A: The "Visual Engineering" Niche (High CPM)**

* **Concept:** Instead of dry tutorials, use your math and ML background to create "visual essays" that explain complex concepts intuitively.
* **Why it works:** There is a massive demand for people who can bridge the gap between high-level academic math (like Calculus or Probability) and practical application (like AI).
* **Video Ideas:**
* *"How Python actually 'hears' audio"* (Explaining MFCC/Spectrograms visually).
* *"The Math that makes Neural Networks learn"* (Visualizing backpropagation).
* *"Why your Minecraft server is lagging: A technical deep dive"* (Explaining server ticks and RAM allocation).



#### **Path B: The "Devlog" & Project Journey**

* **Concept:** "Building in Public." Document the struggle and success of building complex projects. This allows you to show off your Python/CNN skills without needing to be a "teacher."
* **Why it works:** It builds a parasocial connection. Viewers root for you to succeed.
* **Video Ideas:**
* *"I built an AI to detect plant diseases (and it failed 50 times)"* – This frames a technical project as a story with conflict and resolution.
* *"Coding a Speech Recognition model from scratch"* – Show the trial and error, not just the finished code.



#### **Path C: Technical Gaming (The "Mechanic" Niche)**

* **Concept:** Deep dives into the code/logic of games rather than just playing them.
* **Why it works:** Gaming is saturated, but *technical* gaming analysis is evergreen.
* **Video Ideas:**
* *"How Minecraft generates terrain (The Math of Perlin Noise)."*
* *"Automating a Minecraft Server with Python scripts."*



### 4. Your Launch Checklist

1. **Package First:** Before you record, design the Thumbnail and Title. If you can't make a compelling click for the idea, don't make the video.
2. **The "First 30 Seconds" Rule:** Do not introduce yourself ("Hi, I'm Huy"). Start immediately with the promise of the video: *"In this video, we are going to fix X by doing Y..."*
3. **Evergreen vs. Trending:** For your first 5 videos, focus on **Evergreen** topics (things people will search for in 2 years, like "How to use Pandas for X") rather than Trending topics (news), as this builds a base of consistent views.
